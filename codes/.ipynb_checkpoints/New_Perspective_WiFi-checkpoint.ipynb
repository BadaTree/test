```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy
import time
import os
```

```python
class wifimodel():
    def __init__(self, location):
        self.train_dir = f'../data/{location}/train/'
        self.test_dir = f'../data/{location}/test/'
        self.model_df = pd.DataFrame()
        self.test_df = pd.DataFrame()

        self.fil_model_df = pd.DataFrame()
        self.fil_test_df = pd.DataFrame()

        self.SSID_list = []
        
        self.test_x_list = []
        self.test_y_list = []
        
        self.maxarea = 0
        self.maxarea2 = 0
        self.refwifi = []

        self.rssi_thres = 0
        self.range_num = 0
        self.range_num2 = 0

        file_list = os.listdir(self.train_dir)
        for file in file_list:
            df = pd.read_csv(self.train_dir + file,
                             sep="\t", engine='python', encoding="UTF-8", header=None)
            df = pd.DataFrame(
                {"cnt": list(df.iloc[:, 0]), "x": list(df.iloc[:, 1]), "y": list(df.iloc[:, 2]),
                 "ADDR": list(df.iloc[:, 3]),
                 "SSID": list(df.iloc[:, 4]),
                 "RSSI": list(df.iloc[:, 5])})
            self.model_df = pd.concat([self.model_df, df])

        file_list = os.listdir(self.test_dir)
        for file in file_list:
            df = pd.read_csv(self.test_dir + file,
                             sep="\t", engine='python', encoding="UTF-8", header=None)
            df = pd.DataFrame(
                {"cnt": list(df.iloc[:, 0]), "x": list(df.iloc[:, 1]), "y": list(df.iloc[:, 2]),
                 "ADDR": list(df.iloc[:, 3]),
                 "SSID": list(df.iloc[:, 4]),
                 "RSSI": list(df.iloc[:, 5])})
            self.test_df = pd.concat([self.test_df, df])
        
        self.mag_df = pd.read_csv(f'../mag/{location}/magx.txt', sep="\t", engine='python', encoding="cp949", header=None)

        self.ref_mag_df = copy.deepcopy(self.mag_df)
        
        for i in range(self.mag_df.shape[0]):
            for j in range(self.mag_df.shape[1]):
                if ((i % 6 == 0 ) and (j % 6 == 0) and (self.mag_df.iloc[i, j] != 0.0)):
                    self.maxarea += 1
        
        self.maxarea2 = sum(self.ref_mag_df[self.ref_mag_df != 0.0].count())
        self.mag_df[self.mag_df != 0.0] = np.nan

        self.maxX = self.mag_df.shape[0]
        self.maxY = self.mag_df.shape[1]
    
    def cal_area(self, posx, posy):
        if(min(posx) < 0):
            min_x = 0
        else:
            min_x = min(posx)
        if(min(posy) < 0):
            min_y = 0
        else:
            min_y = min(posx)

        temp = self.ref_mag_df.iloc[int(min_x): int(max(posx)), int(min_y) : int(max(posy))]
        return sum(temp[temp != 0.0].count())
    
    def cal_min_max(self, posx, posy):
        return [min(posx), max(posx), min(posy), max(posy)]
    
    def check_in_range(self, x, y, range_list):
        check = (range_list[0] - 12 <= x <= range_list[1] + 12) and \
        (range_list[2] - 12 <= y <= range_list[3] + 12)
        return check
    
    def define_range(self, range_num, rssi_range_num):
        self.range_num = range_num
        self.rssi_range_num = rssi_range_num
    
    def define_cov(self, cov_1, cov_2, cov_3, rssi_thres, tot_cov, tot_rssi_thres):
        self.cov_1 = cov_1
        self.cov_2 = cov_2
        self.cov_3 = cov_3
        self.cov_rssi = rssi_thres
        self.tot_cov = tot_cov
        self.tot_cov_rssi = tot_rssi_thres
        
    def create_refwifi(self, rssi_thres):
        self.rssi_thres = rssi_thres
        
        fil_model_df = self.model_df.loc[self.model_df['RSSI'] >= self.rssi_thres]
        self.fil_test_df = self.test_df.loc[self.test_df['RSSI'] >= self.rssi_thres]

        self.SSID_list = list(fil_model_df['SSID'].unique())
        self.test_x_list = self.fil_test_df['x'].unique()
        self.test_y_list = self.fil_test_df['y'].unique()
        
        self.refwifi = np.zeros((self.maxX, self.maxY, len(self.SSID_list)), dtype = np.int64)
        self.rssiwifi = np.zeros((self.maxX, self.maxY, len(self.SSID_list)), dtype = np.int64)
        
        self.rssi_max = np.max(fil_model_df['RSSI'])
        for i in range(fil_model_df.shape[0]):
            posx = fil_model_df.iloc[i, 1]
            posy = fil_model_df.iloc[i, 2]
            ssid = fil_model_df.iloc[i, 4]
            rssi = fil_model_df.iloc[i, 5]
            if(0 <= posx < self.maxX) and (0 <= posy < self.maxY):
                self.refwifi[int(posx)][int(posy)][self.SSID_list.index(ssid)] = 1
                self.rssiwifi[int(posx)][int(posy)][self.SSID_list.index(ssid)] = rssi
    
    def create_coverage(self, rssi_thres):
        fil_model_df = self.model_df.loc[self.model_df['RSSI'] >= rssi_thres]
        
        self.Cov_SSID_list = list(fil_model_df['SSID'].unique())
        self.Cov_ADDR_list = list(fil_model_df['ADDR'].unique())
        self.Cov_list = []
        self.Min_Max_list = []
        
        self.test_x_list = self.test_df['x'].unique()
        self.test_y_list = self.test_df['y'].unique()
        
        for i in range(len(self.Cov_SSID_list)):
            pos_x = list(fil_model_df.loc[fil_model_df['SSID'] == self.Cov_SSID_list[i]]['x'])
            pos_y = list(fil_model_df.loc[fil_model_df['SSID'] == self.Cov_SSID_list[i]]['y'])
            area = self.cal_area(pos_x, pos_y)
            self.Cov_list.append(area / self.maxarea2)
            self.Min_Max_list.append(self.cal_min_max(pos_x, pos_y))
    
    def example(self, percent):
        self.tot_Cov_SSID_list = list(self.model_df['SSID'].unique())
        self.tot_Cov_list = []
        self.tot_Min_Max_list = []
        
        for i in range(len(self.tot_Cov_SSID_list)):
            rssi = list(self.model_df.loc[self.model_df['SSID'] == self.tot_Cov_SSID_list[i]]['RSSI'])
            rssi.sort(reverse= True)
            rssi_lim = rssi[int(len(rssi) * percent)]
            pos_x = list(self.model_df.loc[(self.model_df['SSID'] == self.tot_Cov_SSID_list[i]) & \
                                          (self.model_df['RSSI'] >= rssi_lim)]['x'])
            pos_y = list(self.model_df.loc[(self.model_df['SSID'] == self.tot_Cov_SSID_list[i]) & \
                                          (self.model_df['RSSI'] >= rssi_lim)]['y'])
            
            area = self.cal_area(pos_x, pos_y)
            self.tot_Cov_list.append(area / self.maxarea2)
            self.tot_Min_Max_list.append(self.cal_min_max(pos_x, pos_y))
    
    def create_tot_coverage(self):
        self.tot_Cov_SSID_list = list(self.model_df['SSID'].unique())
        self.tot_Cov_list = []
        self.tot_Min_Max_list = []
        
        for i in range(len(self.tot_Cov_SSID_list)):
            rssi = list(self.model_df.loc[self.model_df['SSID'] == self.tot_Cov_SSID_list[i]]['RSSI'])
            
            pos_x = list(self.model_df.loc[self.model_df['SSID'] == self.tot_Cov_SSID_list[i]]['x'])
            pos_y = list(self.model_df.loc[self.model_df['SSID'] == self.tot_Cov_SSID_list[i]]['y'])
            
            area = self.cal_area(pos_x, pos_y)
            self.tot_Cov_list.append(area / self.maxarea2)
            self.tot_Min_Max_list.append(self.cal_min_max(pos_x, pos_y))
    
    def test_one(self, idx):
        tot_cnt = 0
        cnt = 0
        self.area_list = []
        coords_list = []
        
        self.test_df['x_y'] = self.test_df['x'].astype('str') + "\t" + self.test_df['y'].astype('str')
        test_list = list(self.test_df['x_y'].unique())[idx].split('\t')
        
        x = float(test_list[0])
        y = float(test_list[1])
        
        compare_df = self.test_df.loc[(self.test_df["x"] == x) & (self.test_df["y"] == y)]
        
        
        range_df = []
        tot_range_df = []
        
        range_df1 = []
        range_df2 = []
        range_df3 = []
        compare_df = self.test_df.loc[(self.test_df["x"] == x) & (self.test_df["y"] == y)]
        if compare_df.empty == False:
            for i in range(compare_df.shape[0]):
                ssid = compare_df.iloc[i, 4]
                rssi = compare_df.iloc[i, 5]
                
                if ssid in self.tot_Cov_SSID_list:
                    if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.tot_cov and rssi >= self.tot_cov_rssi:
                        min_x = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][0]
                        max_x = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][1]
                        min_y = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][2]
                        max_y = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][3]
                        tot_range_df.append([min_x, max_x, min_y, max_y])
            if pd.DataFrame(tot_range_df).empty == False:
                range_df = pd.DataFrame(tot_range_df)
                min_x = min(range_df.iloc[:, 0])
                max_x = max(range_df.iloc[:, 1])
                min_y = min(range_df.iloc[:, 2])
                max_y = max(range_df.iloc[:, 3])
                
                return x, y, [min_x, max_x, min_y, max_y]
            
            else:
                for i in range(compare_df.shape[0]):
                    ssid = compare_df.iloc[i, 4]
                    rssi = compare_df.iloc[i, 5]
                    if ssid in self.Cov_SSID_list:

                        if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_1 and rssi >= self.cov_rssi:
                            min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                            max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                            min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                            max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                            print(self.Cov_list[self.Cov_SSID_list.index(ssid)])
                            print([min_x, max_x, min_y, max_y])
                            range_df1.append([min_x, max_x, min_y, max_y])

                        if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_2 and rssi >= self.cov_rssi:
                            min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                            max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                            min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                            max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                            range_df2.append([min_x, max_x, min_y, max_y])

                        if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_3 and rssi >= self.cov_rssi:
                            min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                            max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                            min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                            max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                            range_df3.append([min_x, max_x, min_y, max_y])


                if pd.DataFrame(range_df1).empty == False:
                    range_df = pd.DataFrame(range_df1)
                elif pd.DataFrame(range_df2).empty == False:
                    range_df = pd.DataFrame(range_df2)
                elif pd.DataFrame(range_df3).empty == False:
                    range_df = pd.DataFrame(range_df3)
                else:
                    range_df = pd.DataFrame([])

                print(range_df)

                if range_df.empty == False:
                    min_x = min(range_df.iloc[:, 0])
                    max_x = max(range_df.iloc[:, 1])
                    min_y = min(range_df.iloc[:, 2])
                    max_y = max(range_df.iloc[:, 3])

                    return x, y, [min_x, max_x, min_y, max_y]
                else:
                    return x, y, [0, self.maxX, 0, self.maxY]
    
    def only_test_cov(self):
        tot_cnt = 0
        cnt = 0
        no_cov_cnt = 0
        cov_fail_cnt = 0
        self.area_list = []
        fail_list = []
        succ_list = []
        coords_list = []

        for x in self.test_x_list:
            for y in self.test_y_list:
                range_df = []
                tot_range_df = []
                range_df1 = []
                range_df2 = []
                range_df3 = []
                compare_df = self.test_df.loc[(self.test_df["x"] == x) & (self.test_df["y"] == y)]
                if compare_df.empty == False:
                    for i in range(compare_df.shape[0]):
                        ssid = compare_df.iloc[i, 4]
                        rssi = compare_df.iloc[i, 5]

                        if ssid in self.tot_Cov_SSID_list:
                            if self.tot_Cov_list[self.tot_Cov_SSID_list.index(ssid)] <= self.tot_cov and rssi >= self.tot_cov_rssi:
                                min_x = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][0]
                                max_x = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][1]
                                min_y = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][2]
                                max_y = self.tot_Min_Max_list[self.tot_Cov_SSID_list.index(ssid)][3]
                                tot_range_df.append([min_x, max_x, min_y, max_y])
                    if pd.DataFrame(tot_range_df).empty == False:
                        range_df = pd.DataFrame(tot_range_df)
                        min_x = min(range_df.iloc[:, 0])
                        max_x = max(range_df.iloc[:, 1])
                        min_y = min(range_df.iloc[:, 2])
                        max_y = max(range_df.iloc[:, 3])
                        if self.check_in_range(x, y, [min_x, max_x, min_y, max_y]):
                            cnt += 1
                            self.area_list.append(((max_x - min_x) * (max_y - min_y)) / (self.maxX * self.maxY))
                            succ_list.append([x, y, [min_x, max_x, min_y, max_y]])
                        else:
                            cov_fail_cnt += 1
                            fail_list.append([x, y, [min_x, max_x, min_y, max_y]])
                    else:
                        for i in range(compare_df.shape[0]):
                            ssid = compare_df.iloc[i, 4]
                            rssi = compare_df.iloc[i, 5]
                            if ssid in self.Cov_SSID_list:

                                if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_1 and rssi >= self.cov_rssi:
                                    min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                                    max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                                    min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                                    max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                                    range_df1.append([min_x, max_x, min_y, max_y])

                                if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_2 and rssi >= self.cov_rssi:
                                    min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                                    max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                                    min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                                    max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                                    range_df2.append([min_x, max_x, min_y, max_y])

                                if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_3 and rssi >= self.cov_rssi:
                                    min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                                    max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                                    min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                                    max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                                    range_df3.append([min_x, max_x, min_y, max_y])

                        if pd.DataFrame(range_df1).empty == False:
                            range_df = pd.DataFrame(range_df1)
                        elif pd.DataFrame(range_df2).empty == False:
                            range_df = pd.DataFrame(range_df2)
                        elif pd.DataFrame(range_df3).empty == False:
                            range_df = pd.DataFrame(range_df3)
                        else:
                            range_df = pd.DataFrame([])


                        if range_df.empty == False:
                            min_x = min(range_df.iloc[:, 0])
                            max_x = max(range_df.iloc[:, 1])
                            min_y = min(range_df.iloc[:, 2])
                            max_y = max(range_df.iloc[:, 3])

                            if self.check_in_range(x, y, [min_x, max_x, min_y, max_y]):
                                cnt += 1
                                self.area_list.append(((max_x - min_x) * (max_y - min_y)) / (self.maxX * self.maxY))
                                succ_list.append([x, y, [min_x, max_x, min_y, max_y]])
                            else:
                                cov_fail_cnt += 1
                                fail_list.append([x, y, [min_x, max_x, min_y, max_y]])
                        else:
                            no_cov_cnt += 1
                    tot_cnt += 1
                    if (tot_cnt % 100 == 0):
                        print(f"현재 진행 상황 : {tot_cnt} 회")
                        print(f"성공 횟수 : {cnt}회")
                        print(f"실패 횟수 : {cov_fail_cnt}회")
                        print(f'COV 사용 못한 횟수 : {no_cov_cnt}회')
        
        
        print(f"test 횟수 : {tot_cnt}, 성공 횟수 : {cnt}, 실패 횟수 : {cov_fail_cnt}, COV 사용 못한 횟수 : {no_cov_cnt}")
        print(cnt * 100 / tot_cnt)
        print(np.average(self.area_list))
        
        return cnt / tot_cnt, np.average(self.area_list), succ_list, fail_list
    
    def test_all(self):
        tot_cnt = 0
        cnt = 0
        cov_fail_cnt = 0
        self.area_list = []
        fail_list = []
        coords_list = []

        for x in self.test_x_list:
            for y in self.test_y_list:
                compare_df = self.test_df.loc[(self.test_df["x"] == x) & (self.test_df["y"] == y)]
                range_df = []

                range_df1 = []
                range_df2 = []
                range_df3 = []
                if compare_df.empty == False:
                    for i in range(compare_df.shape[0]):
                        ssid = compare_df.iloc[i, 4]
                        rssi = compare_df.iloc[i, 5]
                        if ssid in self.Cov_SSID_list:
                            
                            if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_1 and rssi >= self.cov_rssi:
                                min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                                max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                                min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                                max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                                range_df1.append([min_x, max_x, min_y, max_y])
                            
                            if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_2 and rssi >= self.cov_rssi:
                                min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                                max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                                min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                                max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                                range_df2.append([min_x, max_x, min_y, max_y])
                            
                            if self.Cov_list[self.Cov_SSID_list.index(ssid)] <= self.cov_3 and rssi >= self.cov_rssi:
                                min_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][0]
                                max_x = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][1]
                                min_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][2]
                                max_y = self.Min_Max_list[self.Cov_SSID_list.index(ssid)][3]
                                range_df3.append([min_x, max_x, min_y, max_y])
                    
                    if pd.DataFrame(range_df1).empty == False:
                        range_df = pd.DataFrame(range_df1)
                    elif pd.DataFrame(range_df2).empty == False:
                        range_df = pd.DataFrame(range_df2)
                    elif pd.DataFrame(range_df3).empty == False:
                        range_df = pd.DataFrame(range_df3)
                    else:
                        range_df = pd.DataFrame([])
                    
                    
                    if range_df.empty == False:
                        min_x = min(range_df.iloc[:, 0])
                        max_x = max(range_df.iloc[:, 1])
                        min_y = min(range_df.iloc[:, 2])
                        max_y = max(range_df.iloc[:, 3])

                        if self.check_in_range(x, y, [min_x, max_x, min_y, max_y]):
                            cnt += 1
                            self.area_list.append(((max_x - min_x) * (max_y - min_y)) / (self.maxX * self.maxY))
                        else:
                            cov_fail_cnt += 1
                    else:
                        compare_list = np.array([0 for i in range(len(self.SSID_list))])
                        compare_list_rssi = np.array([0 for i in range(len(self.SSID_list))])
                        compare_df = self.fil_test_df.loc[(self.fil_test_df["x"] == x) & (self.fil_test_df["y"] == y)]
                        if compare_df.empty == False:
                            for i in range(compare_df.shape[0]):
                                if (compare_df.iloc[i, 4] in self.SSID_list):
                                    compare_list[self.SSID_list.index(compare_df.iloc[i, 4])] = -1
                                    compare_list_rssi[self.SSID_list.index(compare_df.iloc[i, 4])] = compare_df.iloc[i, 5]

                            temp_wifi = self.refwifi - compare_list
                            temp_wifi = np.where(temp_wifi == 1, 0, temp_wifi)

                            temp_rssi_wifi = abs(self.rssiwifi - compare_list_rssi)
                            temp_rssi_wifi = np.where(abs(temp_rssi_wifi) >= abs(self.rssi_max), 0, temp_rssi_wifi)

                            sum_temp = temp_wifi.sum(axis=2)

                            sum_temp_for_rssi = copy.deepcopy(sum_temp)

                            sum_temp_for_rssi = np.where(sum_temp_for_rssi == 0, 0.00001, sum_temp)
                            sum_rssi_temp = temp_rssi_wifi.sum(axis=2)

                            sum_rssi_temp = sum_rssi_temp / sum_temp_for_rssi

                            sum_np = np.unique(sum_temp.flatten())[:: -1]

                            if len(sum_np) <= self.range_num:
                                range_idx = len(sum_np) - 1
                            else:
                                range_idx = self.range_num

                            if range_idx == -1:
                                rangeval = 0
                            else:
                                rangeval = np.unique(sum_np)[:: - 1][range_idx]

                            sum_rssi_np = np.unique(sum_rssi_temp[np.where(sum_temp >= rangeval)].flatten())

                            if len(sum_rssi_np) <= self.rssi_range_num:
                                range_idx = len(sum_rssi_np) - 1
                            else:
                                range_idx = self.rssi_range_num

                            if range_idx == -1:
                                rssi_rangeval = 0
                            else:
                                rssi_rangeval = np.unique(sum_rssi_np)[range_idx]

                            coords = np.where((sum_temp >= rangeval) & (sum_rssi_temp <= rssi_rangeval))

                            if (min(coords[0]) - 6 <= x <= max(coords[0]) + 6 and min(coords[1]) - 6 <= y <= max(coords[1]) + 6):
                                cnt += 1
                                coords_list.append([x, y, coords])
                                self.area_list.append((max(coords[0]) - min(coords[0])) * (max(coords[1]) - min(coords[1]))
                                                      / (self.maxX * self.maxY))
                            else:
                                fail_list.append([x, y, coords])

                    tot_cnt += 1

                    if (tot_cnt % 100 == 0):
                        print(f"현재 진행 상황 : {tot_cnt} 회")
                        print(f"성공 횟수 : {cnt}회")
                        print(f"COV 실패 횟수 : {cov_fail_cnt}회")
        
        
        print(f"test 횟수 : {tot_cnt}, 성공 횟수 : {cnt}")
        print(cnt * 100 / tot_cnt)
        print(np.average(self.area_list))
        
        return cnt / tot_cnt, np.average(self.area_list), fail_list
    
```
