```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy
import os

location = '포스코실험동'

train_dir = f'../data/{location}/train/'
test_dir = f'../data/{location}/test/'

file_list = os.listdir(train_dir)
model_df = pd.DataFrame()
for file in file_list:
    df = pd.read_csv(train_dir + file,
                     sep="\t", engine='python', encoding="UTF-8", header=None)
    df = pd.DataFrame(
        {"cnt": list(df.iloc[:, 0]), "x": list(df.iloc[:, 1]), "y": list(df.iloc[:, 2]), "ADDR": list(df.iloc[:, 3]),
         "SSID": list(df.iloc[:, 4]),
         "RSSI": list(df.iloc[:, 5])})
    model_df = pd.concat([model_df, df])

file_list = os.listdir(test_dir)
test_df = pd.DataFrame()
for file in file_list:
    df = pd.read_csv(test_dir + file,
                     sep="\t", engine='python', encoding="UTF-8", header=None)
    df = pd.DataFrame(
        {"cnt": list(df.iloc[:, 0]), "x": list(df.iloc[:, 1]), "y": list(df.iloc[:, 2]), "ADDR": list(df.iloc[:, 3]),
         "SSID": list(df.iloc[:, 4]),
         "RSSI": list(df.iloc[:, 5])})
    test_df = pd.concat([test_df, df])
```

```python
def cal_area(posx, posy):
    temp = ref_mag_df.iloc[int(min(posx)): int(max(posx)), int(min(posy)) : int(max(posy))]
    return sum(temp[temp != 0.0].count())    
```

```python
refmagfile = '../handmagx.txt'
mag_df = pd.read_csv(refmagfile, sep="\t", engine='python', encoding="cp949", header=None)

maxarea = 0 

for i in range(mag_df.shape[0]):
    for j in range(mag_df.shape[1]):
        if ((i % 6 == 0 ) and (j % 6 == 0) and (mag_df.iloc[i, j] != 0.0)):
            maxarea += 1
```

```python
refmagfile = '../handmagx.txt'
mag_df = pd.read_csv(refmagfile, sep="\t", engine='python', encoding="cp949", header=None)

ref_mag_df = copy.deepcopy(mag_df)

test = sum(ref_mag_df[ref_mag_df != 0.0].count()) / 36

print(maxarea, test)
mag_df[mag_df != 0.0] = np.nan

maxX = mag_df.shape[0]
maxY = mag_df.shape[1]
```

```python
thres = -100
range_val = 4

model_df = model_df.loc[model_df['RSSI'] >= thres]
test_df = test_df.loc[test_df['RSSI'] >= thres]

SSID_list = list(model_df['SSID'].unique())
ADDR_list = list(model_df['ADDR'].unique())

area_list = []
density_list = []
ssid_ftd_zero = []
freq_list = []
rssi_list = []
rssi_mean = []
rssi_std = []
```

```python
for i in range(len(SSID_list)):
    pos_x = list(model_df.loc[model_df['SSID'] == SSID_list[i]]['x'])
    pos_y = list(model_df.loc[model_df['SSID'] == SSID_list[i]]['y'])
    rssi = list(model_df.loc[model_df['SSID'] == SSID_list[i]]['RSSI'])
    area = cal_area(pos_x, pos_y)
    if area != 0.0:
        ssid_ftd_zero.append(SSID_list[i])
        area_list.append(len(pos_x)/ maxarea)
        density_list.append(36 * len(pos_x) / area)
        freq_list.append(len(pos_x))
        rssi_list.append(rssi)
        rssi_mean.append(np.mean(rssi))
        rssi_std.append(np.std(rssi))
```

```python
def draw_dist(ssid, df, divider):
    
    rssi =  list(model_df.loc[model_df['SSID'] == ssid]['RSSI'])
    rssi_bins = np.arange(np.min(rssi), np.max(rssi), \
                         (np.max(rssi) - np.min(rssi)) / divider)
    ssid_df = copy.deepcopy(model_df.loc[model_df['SSID'] == ssid])
    ssid_df['RSSI_digit'] = np.digitize(ssid_df['RSSI'], rssi_bins)
    
    plt.figure(figsize = (8, 4))
    plt.xlabel('RSSI')
    plt.ylabel('Number Of Reference Points')
    plt.hist(ssid_df['RSSI'], rssi_bins)
    plt.show()
        
    plt.figure(figsize = (18, 14))
    plt.imshow(mag_df, cmap='jet', interpolation='none')
    plt.scatter(ssid_df.y, ssid_df.x, c=ssid_df.RSSI, cmap = plt.cm.Reds, s = 10)
    plt.show()
    
    for i in range(len(rssi_bins)):
        print(rssi_bins[i])
        temp_df = ssid_df.loc[ssid_df['RSSI_digit'] == i + 1]
        plt.figure(figsize = (18, 14))
        plt.imshow(mag_df, cmap='jet', interpolation='none')
        plt.scatter(temp_df.y, temp_df.x, c='r', s = 20)
        plt.show()
    
    
#     pos_x = list(model_df.loc[model_df['SSID'] == ssid]['x'])
#     pos_y = list(model_df.loc[model_df['SSID'] == ssid]['y'])
```

```python
past_df['x_y'] = past_df['x'].astype('str') + "\t" + past_df['y'].astype('str')
test_list = list(past_df['x_y'].unique())[0].split('\t')
x = float(test_list[0])
y = float(test_list[1])
SSID_unq = past_df.loc[(past_df['x'] == x) & (past_df['y'] == y)]['SSID'].unique()
for j in range(10, 20):#len(SSID_unq)):
    step_list = []
    walk_list = []
    print(SSID_unq[j])
    for idx in range(20):
        test_list = list(past_df['x_y'].unique())[idx].split('\t')
        x = float(test_list[0])
        y = float(test_list[1])
        step_rssi = np.mean(list(model_df.loc[(model_df['SSID'] == SSID_unq[j]) 
                                      & (model_df['x'] == x)
                                                     &(model_df['y'] == y)]['RSSI']))
        walk_rssi = np.mean(list(past_df.loc[(past_df['SSID'] == SSID_unq[j]) & (past_df['x'] == x)
                                                     &(past_df['y'] == y)]['RSSI']))
        step_list.append(step_rssi)
        walk_list.append(walk_rssi)
    plt.figure(figsize = (8, 5))
    plt.plot(step_list, label = 'step by step')
    plt.plot(walk_list, label = 'walking')
    plt.legend(loc='best')
    plt.show()
```
